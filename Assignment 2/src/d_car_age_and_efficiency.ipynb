```python
import os
import pandas as pd

df=pd.read_csv('/content/missing_values_replaced.csv')
#obtaining current age of the car and storing it under the new column 'Age'
df['Age']=2024-df['Year']
#calculating efficiency and storing it under the new column 'Efficiency'
df['Efficiency']=df['Mileage']/df['Power']

df.head()
```
