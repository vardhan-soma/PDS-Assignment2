**a) Look for the missing values in all the columns and either impute them (replace with mean,
median, or mode) or drop them. Justify your action for this task.**

```python
import os
import pandas as pd
# reading csv file by using inbuilt pandas function
df=pd.read_csv('/content/units_removed.csv')
df
```

```python
df.info() #returns the columns along with their data types
```

```python
df.describe() #gives a brief idea about the dataset
```

```python
df.isna() #to check null values
```

```python
# obtaining the count of missing values in each column
missing_values = df.isnull().sum()
print(missing_values)
```

```python
#to obtain the only the columns containg missing values

missing = missing_values[missing_values>0]
print(missing)
```

```python
df['Mileage'].fillna(df['Mileage'].median(), inplace=True)  #replacing null values with median
df['Engine'].fillna(df['Engine'].mean(), inplace=True)      #replacing null values with mean
df['Power'].fillna(df['Power'].median(), inplace=True)      #replacing null values with mean
df['Seats'].fillna(df['Seats'].mean(), inplace=True)        #replacing null values with mean
df.drop('New_Price', axis=1, inplace=True)                  #dropping the New_Price column
new_values = df.isnull().sum()
print(new_values)
```

```python
df.to_csv('/content/missing_values_replaced.csv')
```
