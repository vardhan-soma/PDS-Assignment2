```python
import os
import pandas as pd

df=pd.read_csv('/content/age_and_efficiency.csv')
df.head()
```

```python
#selecting a few columns

selected_columns = df[['Name','Year','Kilometers_Driven','Fuel_Type','Transmission','Age']]

selected_columns.head()
```

```python
#filter

transmission = df[df['Transmission']=='Automatic']
transmission.head()
```

```python
#filter

km = df[df['Kilometers_Driven']<50000]
km.head()
```

```python
#renaming the columns Location, Kilometers driven and transmision
df.rename(columns={'Location':'Manufactured Location','Kilometers_Driven':'Mileometer(in kms)','Transmission':'Mode'}, inplace=True)
df.head()
```

```python
#adding one more feature which is equivalent to mutate in R

df['Avg_kms_per_year']=df['Mileometer(in kms)']/df['Age']
df.head()
```

```python
#arrange

sort=df.sort_values(by='Year', ascending=False)
sort
```

```python
less_kms=df.sort_values(by='Mileometer(in kms)', ascending=True)
less_kms
```

```python
#grouping the data by the mode of the vehicle and by the mean price
grouped_data= df.groupby('Mode')['Price'].mean()
grouped_data
```

```python
#grouping the data by the owner type and by the mean price
grouped_data= df.groupby('Owner_Type')['Price'].mean()
grouped_data
```

```python
#grouping the data by fuel type of the vehicle and by the mean of mileage
grouped_data= df.groupby('Fuel_Type')['Mileage'].mean()
grouped_data
```
